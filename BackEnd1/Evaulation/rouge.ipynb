```python
# install dulu kalau belum
# pip install rouge-score

from rouge_score import rouge_scorer

# definisikan prediksi dan referensi
predictions = [
    "Yesus lahir di Betlehem.",
    "Musa memimpin bangsa Israel keluar dari Mesir.",
    "Daud mengalahkan Goliat dengan batu."
]

references = [
    "Yesus Kristus lahir di kota Betlehem.",
    "Musa membawa umat Israel keluar dari tanah Mesir.",
    "Raja Daud mengalahkan Goliat menggunakan batu kecil."
]

# inisialisasi ROUGE scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

# hitung skor untuk setiap pasangan
for i, (pred, ref) in enumerate(zip(predictions, references)):
    scores = scorer.score(ref, pred)
    print(f"Contoh {i+1}:")
    for metric, result in scores.items():
        print(f"  {metric}: Precision={result.precision:.3f}, Recall={result.recall:.3f}, F1={result.fmeasure:.3f}")
    print()
```

```python
from nltk.translate.bleu_score import sentence_bleu
from rouge_score import rouge_scorer

ref = ["Yesus Kristus lahir di kota Betlehem"]
pred = "Yesus lahir di Betlehem"

# BLEU
bleu = sentence_bleu([ref[0].split()], pred.split())
print(f"BLEU: {bleu:.3f}")

# ROUGE
scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
scores = scorer.score(ref[0], pred)
print(f"ROUGE-1 F1: {scores['rouge1'].fmeasure:.3f}")
```
